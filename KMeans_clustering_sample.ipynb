```python
from sklearn.cluster import KMeans
```

```python
kme = KMeans(n_clusters=2)
```

```python
kme.fit(college.drop(columns=['Private'],axis=1))
```

```python
kme.cluster_centers_
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(college['Private'])
y = le.transform(college['Private'])
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y,kme.labels_),'\n',classification_report(y,kme.labels_))
```

```python
print('Accuracy_score',accuracy_score(y,kme.labels_))
```
